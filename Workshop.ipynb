## Look for, Create and Remove Missing Data


```python
#Checks for NaNs in the dataframe
patients.isnull().any(axis=0)
```

```python
#Creates a copy of patients and assigns it to variable patients2 (we'll use patients as a backup)
patients2=patients.copy()
```

```python
#Obtains a 5x3 numpy array of random integers
missing_index=np.random.randint(0,len(patients2)-1,size=(5,3))
missing_index
```

```python
#Assigns other characters to the dataframe in data2 using the positions set above
patients2.ix[missing_index[:,0],'sex']=np.nan
patients2.ix[missing_index[:,1],'age']=' '
patients2.ix[missing_index[:,2],'wgt']='.'
```

```python
#Displays the new dataframe
patients2
```

```python
#Summarizes the dataframe
patients2.describe(include='all')
```

```python
#Check data types
patients2.dtypes
```

```python
#Checks whether there are any NaNs in the dataframe
patients2.isnull().any(axis=0)
```

```python
#Replace other characters by NaNs
patients2=patients2.replace(' ',np.nan)
patients2=patients2.replace('.',np.nan)
patients2
```

```python
#Obtains the index where there are NaNs in the dataframe
nan_df_index=np.where(patients2.isnull())
nan_df_index
```

```python
#Displays the length of the first element of the tuple
len(nan_df_index[0])
```

```python
#Removes any lines from the dataframe that contains NaNs 
patients3=patients2.dropna(axis=0,how='any')
patients3
```

```python
#Checks for NaNs
patients3.isnull().any()
```

```python
#Displays lines and column number
patients3.shape
```

```python
#Fill the missing data using linear interpolation
patients3=patients2.interpolate(method='linear')
patients3
```

```python
#Displays lines and column number
patients3.shape
```

## Apply Function, Group and Merge Data


```python
#Displays the first few lines of the dataframe (using the 'original dataframe)
patients.head()
```

```python
#Summarize the dataframe
patients.describe(include='all')
```

```python
#Changes the order of the columns on the dataframe
patients=patients.iloc[:,[0,3,1,2,4,5]]
patients.head()
```

```python
#Obtains the mean of each one of the numerical columns on the dataframe
patients.mean()
```

```python
#Obtains the mean of each one of the numerical columns on the dataframe
patients.iloc[:,2:].apply(np.mean)
```

```python
#Obtains the cumulative sum along the columns
patients.iloc[:,2:].apply(np.cumsum,axis=0).head()
```

```python
#Obtains the difference between the max and min for each one of the columns
patients.iloc[:,2:].apply(lambda x: x.max() - x.min())
```

```python
#Confirms the difference above for the column 'age
patients.age.max()-patients.age.min()
```

```python
#Groups the data by 'sex and obtains the mean of each column for each group
patients.groupby('sex').mean()
```

```python
#Groups the data by 'sex and then smoke and obtains the mean of each column for each group
patients.groupby(['sex','smoke']).std()
```

```python
#Creates an array with three random row indexes from the dataframe and duplicates them
sample_index=patients.index.values[np.random.randint(0,len(patients)-1,size=3)]
sample_index
```

```python
#Creates an array with three elements and duplicates them
new_var=np.array(['T','S','T'])
new_var
```

```python
#Create a new dataframe with two column variables
patients2 = pd.DataFrame({'A':pd.Categorical(sample_index),'B':pd.Categorical(new_var)})
patients2
```

```python
patients2=pd.concat([patients2,patients2])
patients2
```

```python
#Displays the signature for the function merge
?pd.merge
```

```python
#Merges the data on the original dataframe (left dataframe) with the new dataframe (right dataframe)
patients3=pd.merge(patients,patients2,how='inner',left_index=True,right_on='A')
patients3
```

```python
#Writes the new merged dataframe to csv and excel files
patients3.to_csv('newData.csv')
patients3.to_excel('newData.xlsx', sheet_name='Sheet1')
```

```python
#Displays the histograms of 'weight grouped by 'smoke in the same plot
plt.rcParams.update({'font.size': 20})
plt.figure()
patients.groupby('smoke').wgt.hist(alpha=0.5)
plt.legend(labels=['Smoke:No','Smoke:Yes'],loc='best')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.grid('off')
```
